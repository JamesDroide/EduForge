```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configuración visual
plt.style.use("ggplot")
sns.set(font_scale=1.1)
```

```python
#Carga de datos
data_path = r'C:\Users\james\OneDrive\Documentos\PREGRADO UPAO\CICLO 9\Tesis I\Proyecto\BackEnds\EduForge\data\student_data.csv'
df = pd.read_csv(data_path)
```

```python
#Inspeccion inicial
print("Dimensiones:", df.shape)
print("\nPrimeras filas:")
df.head()
```

```python
# Verificamos el tipo de datos y valores nulos
print("\nTipos de datos:")
df.info()

print("\nValores nulos:")
df.isnull().sum()
```

```python
# Verificamos la stadísticas descriptivas
df.describe()
```

```python
#Análisis de distribución de variables numéricas
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribución de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.tight_layout()
    plt.show()
```

```python
 # Boxplots por género
if 'sex' in df.columns:
    for col in num_cols:
        plt.figure(figsize=(8, 4))
        sns.boxplot(data=df, x='sex', y=col)
        plt.title(f'{col} según Sexo')
        plt.tight_layout()
        plt.show()
```

```python
# Correlación entre variables
plt.figure(figsize=(12, 8))
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de Correlación")
plt.tight_layout()
plt.show()
```

```python
# Valores atípicos (outliers visuales)
for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df, y=col)
    plt.title(f'Detección de outliers en {col}')
    plt.tight_layout()
    plt.show()
```

```python
# Visualización por categoría: Acceso a recursos
if 'Access_to_Resources' in df.columns:
    for col in num_cols:
        plt.figure(figsize=(8, 4))
        sns.boxplot(data=df, x='Access_to_Resources', y=col)
        plt.title(f'{col} según Acceso a recursos')
        plt.xlabel("Acceso a recursos")
        plt.ylabel(col)
        plt.tight_layout()
        plt.show()
```
