```python
import sys
import os
import seaborn as sns
import matplotlib.pyplot as plt

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
from notebooks.data_1.data_preprocessing import cargar_y_preprocesar_datos
```

```python
#Carga de datos
data_path = r'C:\Users\james\OneDrive\Documentos\PREGRADO UPAO\CICLO 9\Tesis I\BackEnds\EduForgeF\EduForge\data\StudentPerformanceFactors.csv'
df = cargar_y_preprocesar_datos(data_path, normalizar=True)
```

```python
# Selección de columnas numéricas
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Graficar distribuciones
for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribución de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.tight_layout()
    plt.show()
```

```python
# Matriz de correlación
plt.figure(figsize=(12, 8))
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de Correlación")
plt.tight_layout()
plt.show()
```

```python
# Boxplot para detección de outliers
for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df, y=col)
    plt.title(f'Detección de outliers en {col}')
    plt.tight_layout()
    plt.show()
```
